```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

def preprocess_data(df, label_col="label"):
    """
    Preprocess the DataFrame:
    1. Separate label from features.
    2. Replace inf/-inf with NaN.
    3. Fill NaNs with 0.
    4. Remove columns with zero variance.
    5. (Optional) Scale features.
    Returns X (features) and y (labels).
    """
    # Separate label and features
    y = df[label_col].values
    # Drop non-numeric columns and the label column
    X = df.drop(columns=[label_col])
    X = X.select_dtypes(include=[np.number])  # Keep only numeric columns

    # Replace inf with NaN
    X = X.replace([np.inf, -np.inf], np.nan)

    # Fill NaNs with 0 (could also do dropna, or fill with mean, etc.)
    X = X.fillna(0)

    # Remove columns with zero variance
    variances = X.var(axis=0)
    cols_to_keep = variances[variances != 0].index
    X = X[cols_to_keep]

    # Optional: scale features (StandardScaler, MinMaxScaler, etc.)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, y

def incremental_learning(X, y, batch_size=500):
    """
    Perform incremental training on X, y using stratified sampling:
    1. Split into train/test (stratified).
    2. Use StratifiedKFold to create ~batch_size folds on the training set.
    3. Train an SVM on cumulative folds, evaluate on test set, and track accuracy.
    4. Plot accuracy vs. #training samples.
    """
    # Train/test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )
    print(f"Training set: {X_train.shape[0]} samples, Test set: {X_test.shape[0]} samples.")

    n_train = X_train.shape[0]
    n_batches = n_train // batch_size
    print(f"Will do incremental learning in {n_batches} batches of ~{batch_size} samples each.")

    skf = StratifiedKFold(n_splits=n_batches, shuffle=True, random_state=42)
    cumulative_indices = np.array([], dtype=int)

    sample_sizes = []
    accuracies = []

    fold_number = 1
    for fold_indices, _ in skf.split(X_train, y_train):
        # Add new fold indices
        cumulative_indices = np.concatenate([cumulative_indices, fold_indices])
        cumulative_indices = np.unique(cumulative_indices)

        X_subset = X_train[cumulative_indices]
        y_subset = y_train[cumulative_indices]

        # Check how many unique classes in y_subset
        unique_classes = np.unique(y_subset)
        if len(unique_classes) < 2:
            print(f"Skipping fold {fold_number}: only {len(unique_classes)} class present in training subset.")
            fold_number += 1
            continue

        clf = SVC(kernel='linear', random_state=42)
        clf.fit(X_subset, y_subset)

        y_pred = clf.predict(X_test)
        acc = accuracy_score(y_test, y_pred)

        sample_sizes.append(len(cumulative_indices))
        accuracies.append(acc)

        print(f"Batch {fold_number}: {len(cumulative_indices)} samples -> Test Accuracy: {acc:.4f}")
        fold_number += 1

    # Plot learning curve
    plt.figure(figsize=(8, 6))
    plt.plot(sample_sizes, accuracies, marker='o')
    plt.title("Incremental Learning Curve (Stratified Sampling)")
    plt.xlabel("Number of Training Samples (Cumulative)")
    plt.ylabel("Test Accuracy")
    plt.grid(True)
    plt.show()

    # Check for plateau
    plateau_threshold = 0.001  # define how small a change is considered a plateau
    for i in range(1, len(accuracies)):
        diff = accuracies[i] - accuracies[i-1]
        if abs(diff) < plateau_threshold:
            print(f"Plateau detected at {sample_sizes[i]} samples (accuracy change = {diff:.4f}).")
            break

if __name__ == "__main__":
    # 1. Load your CSV dataset (which has a 'label' column and many 'feature_x' columns)
    df = pd.read_csv("features.csv")  # Adjust path/filename as needed

    # 2. Preprocess the data
    print("Preprocessing data...")
    X_preprocessed, y_preprocessed = preprocess_data(df, label_col="label")
    print(f"Feature matrix shape after preprocessing: {X_preprocessed.shape}")

    # 3. Perform incremental training with stratified sampling
    print("Performing incremental learning...")
    incremental_learning(X_preprocessed, y_preprocessed, batch_size=500)
```
